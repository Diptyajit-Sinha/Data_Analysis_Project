```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assume dataset 'a' is already loaded

# Find highest and lowest total bill by sex
max_bill_sex = a.groupby("sex")["total_bill"].max()
min_bill_sex = a.groupby("sex")["total_bill"].min()
max_tip_sex = a.groupby("sex")["tip"].max()  
min_tip_sex = a.groupby("sex")["tip"].min()  

# Find highest and lowest total bill & tip by smoker
max_bill_smoker = a.groupby("smoker")[["total_bill", "tip"]].max()
min_bill_smoker = a.groupby("smoker")[["total_bill", "tip"]].min()

fig, axes = plt.subplots(4, 2, figsize=(12, 16))

# --- Row 1: Total Bill by Gender ---
sns.barplot(ax=axes[0, 0], x=max_bill_sex.index, y=max_bill_sex.values, palette="coolwarm") #we are passing index in the x and values in the y of max_bill_sex
axes[0, 0].set_xlabel("Gender")
axes[0, 0].set_ylabel("Highest Total Bill")
axes[0, 0].set_title("Highest Total Bill by Gender")
for container in axes[0, 0].containers:
    axes[0, 0].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3)

sns.barplot(ax=axes[0, 1], x=min_bill_sex.index, y=min_bill_sex.values, palette="dark:#1B60D8")
axes[0, 1].set_xlabel("Gender")
axes[0, 1].set_ylabel("Lowest Total Bill")
axes[0, 1].set_title("Lowest Total Bill by Gender")
for container in axes[0, 1].containers:
    axes[0, 1].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3)

# --- Row 2: Total Bill by Smoker ---
sns.barplot(ax=axes[1, 0], x=max_bill_smoker.index, y=max_bill_smoker["total_bill"], palette="coolwarm")
axes[1, 0].set_xlabel("Smoker")
axes[1, 0].set_ylabel("Highest Total Bill")
axes[1, 0].set_title("Highest Total Bill by Smoker")
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3) #padding is used to add the space between two plot,axes,title

sns.barplot(ax=axes[1, 1], x=min_bill_smoker.index, y=min_bill_smoker["total_bill"], palette="dark:#1B60D8")
axes[1, 1].set_xlabel("Smoker")
axes[1, 1].set_ylabel("Lowest Total Bill")
axes[1, 1].set_title("Lowest Total Bill by Smoker")
for container in axes[1, 1].containers:
    axes[1, 1].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3)

# --- Row 3: Tips by Smoker ---
sns.barplot(ax=axes[2, 0], x=max_bill_smoker.index, y=max_bill_smoker["tip"], palette="coolwarm")
axes[2, 0].set_xlabel("Smoker")
axes[2, 0].set_ylabel("Highest Tip")
axes[2, 0].set_title("Highest Tip by Smoker")
for container in axes[2, 0].containers:
    axes[2, 0].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3)

sns.barplot(ax=axes[2, 1], x=min_bill_smoker.index, y=min_bill_smoker["tip"], palette="dark:#1B60D8")
axes[2, 1].set_xlabel("Smoker")
axes[2, 1].set_ylabel("Lowest Tip")
axes[2, 1].set_title("Lowest Tip by Smoker")
for container in axes[2, 1].containers:
    axes[2, 1].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3)

# --- Row 4: Tips by Gender ---
sns.barplot(ax=axes[3, 0], x=max_tip_sex.index, y=max_tip_sex.values, palette="coolwarm")
axes[3, 0].set_xlabel("Gender")
axes[3, 0].set_ylabel("Highest Tip")
axes[3, 0].set_title("Highest Tip by Gender")
for container in axes[3, 0].containers:
    axes[3, 0].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3)

sns.barplot(ax=axes[3, 1], x=min_tip_sex.index, y=min_tip_sex.values, palette="dark:#1B60D8")
axes[3, 1].set_xlabel("Gender")
axes[3, 1].set_ylabel("Lowest Tip")
axes[3, 1].set_title("Lowest Tip by Gender")
for container in axes[3, 1].containers:
    axes[3, 1].bar_label(container, fmt="%.2f", fontsize=12, fontweight='bold', color='black', padding=3)

# Add text below each graph
labels = [
    f"Highest Total Bills:\n{max_bill_sex.to_string()}",
    f"Lowest Total Bills:\n{min_bill_sex.to_string()}",
    f"Highest Total Bills (Smoker):\n{max_bill_smoker['total_bill'].to_string()}",
    f"Lowest Total Bills (Smoker):\n{min_bill_smoker['total_bill'].to_string()}",
    f"Highest Tips (Smoker):\n{max_bill_smoker['tip'].to_string()}",
    f"Lowest Tips (Smoker):\n{min_bill_smoker['tip'].to_string()}",
    f"Highest Tips (Gender):\n{max_tip_sex.to_string()}",
    f"Lowest Tips (Gender):\n{min_tip_sex.to_string()}"
]

for i, ax in enumerate(axes.flat):
    ax.text(0.5, -0.35, labels[i], fontsize=10, ha="center", va="top", transform=ax.transAxes, bbox=dict(facecolor="lightgray", edgecolor="black", boxstyle="round,pad=0.3"))
#transfor=it is used to positioning inside a subplot
#ha=horizontal alignment
#va=vertical alignment
#bbox=is an parameter which is used to create background box around the text
plt.tight_layout()
plt.show()
```
