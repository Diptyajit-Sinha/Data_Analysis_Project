```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_palette("Set2")

# ----------------------- LOAD DATASET -----------------------
df = pd.read_csv("C:\\Users\\Diptyajit Sinha\\.cache\\kagglehub\\datasets\\datasnaek\\youtube-new\\versions\\115\\CAvideos.csv")  

```

```python
# ----------------------- PREPROCESSING -----------------------
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m', errors='coerce')

# Extract time features
df['year_published'] = df['publish_time'].dt.year
df['month_published'] = df['publish_time'].dt.month
df['day_published'] = df['publish_time'].dt.day
df['hour_published'] = df['publish_time'].dt.hour
df['title_length'] = df['title'].str.len()
df['word_count'] = df['title'].str.split().str.len()

# Fill missing values
df['channel_title'] = df['channel_title'].fillna("Unknown")
df['tags'] = df['tags'].fillna("Unknown")
```

```python
# ----------------------- BASIC INFO -----------------------
print(df.info())
print(df.describe())
print("Total videos:", df.shape[0])
print("Unique channels:", df['channel_title'].nunique())
```
